```python
########################### Sales vs Target Analysis ##############################
```

```python
def plot_sales_vs_target(data, category):
    # Filter the data for the specific category
    category_data = data[data['Category'] == category]
    
    # Set the x-axis positions
    x = range(len(category_data))

    # Create a bar plot
    plt.figure(figsize=(12, 6))
    plt.bar(x, category_data['Sales'], width=0.4, label='Sales', align='center')
    plt.bar(x, category_data['Target'], width=0.4, label='Target', align='edge')

    # Add labels and title
    plt.xlabel('Month-Year')
    plt.ylabel('Amount')
    plt.title(f'Sales vs Target for {category}')
    plt.xticks(x, category_data['Month-Year'], rotation=45)
    plt.legend(loc='upper left')

    # Adjust layout and show the plot
    plt.tight_layout()
    plt.show()
```

```python
for cate in merged_sales_target['Category'].unique():
    plot_sales_vs_target(merged_sales_target, cate)
```
